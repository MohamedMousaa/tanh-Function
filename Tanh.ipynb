```python
import math

class NeuralNetwork:
    def __init__(self, input_numbers, hidden_layers_numbers, bias1=None, bias2=None):
        self.input_numbers = input_numbers
        self.hidden_layers_numbers = hidden_layers_numbers
        self.bias1 = bias1
        self.bias2 = bias2
        self.input_weights = []
        self.hidden_weights = []
        self.input_hidden_weight_result = []
        self.TanhValues = []
        self.hidden_output_weight_values_result = []
        self.hidden_output_weight_result = []

    def set_biases(self):
        if self.bias1 is None:
            self.bias1 = float(input("Enter bias1: "))
        if self.bias2 is None:
            self.bias2 = float(input("Enter bias2: "))

    def input_weight_values(self):
        for idx in range(1, self.input_numbers + 1):
            value = float(input(f"Enter Input {idx}: "))
            self.input_weights.append(value)

    def weight_values(self):
        for i in range(1, self.hidden_layers_numbers + 1):
            for j in range(1, self.hidden_layers_numbers + 1):
                value = float(input(f"Enter weight from i{i} to h{j}: "))
                self.hidden_weights.append(value)

    def input_hidden_weight_calculator(self):
        h1 = (self.input_weights[0] * self.hidden_weights[0] + self.bias1) + (self.input_weights[1] * self.hidden_weights[2] + self.bias1)
        h2 = (self.input_weights[0] * self.hidden_weights[1] + self.bias1) + (self.input_weights[1] * self.hidden_weights[3] + self.bias1)
        self.input_hidden_weight_result = [h1, h2]

    def tanh(self):
        tanh1 = (math.exp(self.input_hidden_weight_result[0])-math.exp(-self.input_hidden_weight_result[0]))/(math.exp(self.input_hidden_weight_result[0])+math.exp(-self.input_hidden_weight_result[0]))
        tanh2 = (math.exp(self.input_hidden_weight_result[1])-math.exp(-self.input_hidden_weight_result[1]))/(math.exp(self.input_hidden_weight_result[1])+math.exp(-self.input_hidden_weight_result[1]))
        self.TanhValues = [tanh1, tanh2]

    def hidden_output_weight_values(self):
        for i in range(1, len(self.TanhValues) + 1):
            for j in range(1, len(self.TanhValues) + 1):
                value = float(input(f"Enter weight from h{i} to o{j}: "))
                self.hidden_output_weight_values_result.append(value)

    def hidden_output_weight_calculator(self):
        o1 = (self.TanhValues[0] * self.hidden_output_weight_values_result[0] + self.bias2) + (self.TanhValues[1] * self.hidden_output_weight_values_result[2] + self.bias2)
        o2 = (self.TanhValues[0] * self.hidden_output_weight_values_result[1] + self.bias2) + (self.TanhValues[1] * self.hidden_output_weight_values_result[3] + self.bias2)
        self.hidden_output_weight_result = [o1, o2]

    def run(self):
        self.set_biases()
        self.input_weight_values()
        self.weight_values()
        self.input_hidden_weight_calculator()
        print("Input to Hidden Layer weights:", self.input_hidden_weight_result)
        self.tanh()
        print("Tanh Values:", self.TanhValues)
        self.hidden_output_weight_values()
        self.hidden_output_weight_calculator()
        print("Hidden to Output Layer weights:", self.hidden_output_weight_result)
```

```python
input_numbers = 2
hidden_layers_numbers = 2

nn = NeuralNetwork(input_numbers, hidden_layers_numbers)
nn.run()
```
